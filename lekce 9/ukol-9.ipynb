```python
import pandas
```

```python
import requests

url = "https://raw.githubusercontent.com/andywaltlova/python-1-podzim-2022/master/ukoly/data/temperature.csv"
r = requests.get(url, allow_redirects=True, verify=False)
open("temperature.csv", "wb").write(r.content)
```

```python
temp = pandas.read_csv("temperature.csv", sep = ",")
temp
```

```python
#Vypiš si prvních několik řádků, ať si prohlédneš strukturu tabulky.

temp.head()
```

```python
temp = temp.set_index("City")
temp
```

```python
#Dotaz na měření, která byla provedena v Praze.
temp.loc["Prague"]
```

```python
#Dotaz na měření, ve kterých je teplota (sloupec AvgTemperature) vyšší než 80 stupňů.

temp[temp["AvgTemperature"]>80]
```

```python
#Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

temp.query("AvgTemperature > 60 and Region == 'Europe'")
```

```python
#Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než - 20 stupňů.

temp.query("AvgTemperature > 80 or AvgTemperature < -20")
```

```python
import pytemperature
```

```python
#BONUS
#Nainstaluj si modul pytemperature a zkus si vytvořit nový sloupec, který bude obsahovat průměrnou templotu ve stupních Celsia. Ve svém programu nejprve proveď import modulu pytemperature.

temp ["AvgTemperatureCelsia"] = pytemperature.f2c(temp["AvgTemperature"])
temp
```

```python
#Dotaz na měření, ve kterých je teplota (sloupec AvgTemperatureCelsia) vyšší než 30 stupňů Celsia.

temp [temp["AvgTemperatureCelsia"]>30]
```

```python
#Dotaz na měření, ve kterých je teplota vyšší než 15 stupňů Celsia a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

temp [(temp["AvgTemperatureCelsia"] > 15) & (temp ["Region"] == "Europe" )]
```

```python
#Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 30 stupňů Celsia nebo menší než -10 stupňů. 

temp [(temp["AvgTemperatureCelsia"] > 30) | (temp ["AvgTemperatureCelsia"] < -10)]
```

```python
#the same as previous step but using query

temp.query("AvgTemperatureCelsia > 30 or AvgTemperatureCelsia < -10 ")
```
